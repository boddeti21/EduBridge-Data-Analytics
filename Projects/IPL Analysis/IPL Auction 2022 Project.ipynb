## fillna():

- There are no null values present in our dataset. 
- But if there are any null values present we can fill those places with any other value using the fillna() function of DataFrame.
- There are missing values in State Association, IPL Matches, Previous IPLTeam, Sold Price & New IPL Team.
- There are some categorical features in our dataset.

```python
IPL1=pd.DataFrame(IPL,columns=["Country","Player","State Association","IPL","Sold Price","Base Price",
                               "Previous IPLTeam(s)","2021 Team","Bid","New Franchise","Increase Percentage"])
IPL1
```

```python
IPL1["State Association"].fillna("F",inplace=True)
IPL1["IPL"].fillna(0,inplace=True)
IPL1["Sold Price"].fillna("0 CR",inplace=True)
IPL1["Previous IPLTeam(s)"].fillna("Do No Play",inplace=True)
IPL1["2021 Team"].fillna("DNP",inplace=True)
IPL1["New Franchise"].fillna("Not Selected",inplace=True)
IPL1["Increase Percentage"].fillna(0,inplace=True)
```

```python
IPL1
```

## Converting CR into Lakhs:

```python
def change(item):
    price = item.split()
    if price[1] == "CR":
        return float(price[0])*100
    else:
        return float(price[0])
    
IPL1["Sold Price"] = IPL1["Sold Price"].apply(lambda item: change(item))
```

```python
def Franchise_plot(Franchise):
    Franchise_counts = Franchise.value_counts().sort_values(ascending=True)
    fig, ax = plt.subplots()

    bars = ax.barh(range(len(Franchise_counts)), Franchise_counts, color="purple")
    ax.set_frame_on(False)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.set_yticks(range(len(Franchise_counts)))
    ax.set_yticklabels(Franchise_counts.index)
    
    for i,bar in enumerate(bars):
        value = bar.get_width()
        ax.text(value,bar.get_y() + bar.get_height()/2, '{:,} ({:}%)'.format(int(value), round(value*100/590,2)))
    
Franchise_plot(IPL2["New Franchise"])
plt.show()
```
