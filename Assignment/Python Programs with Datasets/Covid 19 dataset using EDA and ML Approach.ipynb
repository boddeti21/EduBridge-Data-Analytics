## Week of the year in Observation Date

```python
date_wise["Week"] = date_wise.index.weekofyear
```

```python
date_wise
```

```python
week_num = []
week_confirmed = []
week_recovered = []
week_deaths = []
```

```python
w = 1
for i in list(date_wise["Week"].unique()):
    week_confirmed.append(date_wise[date_wise["Week"] == i]["Confirmed"].iloc[-1])
    week_recovered.append(date_wise[date_wise["Week"] == i]["Recovered"].iloc[-1])
    week_deaths.append(date_wise[date_wise["Week"] == i]["Deaths"].iloc[-1])
    week_num.append(w)
    w = w + 1
```

```python
week_num
```

```python
week_confirmed
```

```python
week_recovered
```

```python
week_deaths
```

```python
plt.plot(week_num,week_confirmed)
plt.show()
```

```python
plt.plot(week_num,week_recovered)
plt.show()
```

```python
plt.plot(week_num,week_deaths)
plt.show()
```

```python
plt.plot(week_num,week_confirmed,color="violet",marker="o")
plt.plot(week_num,week_recovered,color="green",marker="x")
plt.plot(week_num,week_deaths,color="black",marker="*")

plt.xlabel("WEEKS",fontsize=15,color="red")
plt.ylabel("PROGRESS IN EACH CASES",fontsize=15,color="red")
plt.title("WEEKLY CASES",fontsize=20,color="blue")

plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize = (15,8))

sns.barplot(x = week_num, y = pd.Series(week_confirmed).diff().fillna(0), ax = ax1)
sns.barplot(x = week_num, y = pd.Series(week_deaths).diff().fillna(0), ax = ax2)

ax1.set_xlabel("WEEK NUMBER")
ax2.set_xlabel("WEEK NUMBER")

ax1.set_ylabel("NUMBER OF CONFIRMED CASES")
ax2.set_ylabel("NUMBER OF DEATH CASES")

ax1.set_title("WEEKLY PROGRESS OF CONFIRMED CASES")
ax2.set_title("WEEKLY PROGRESS OF DEATH CASES")

plt.show()
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize = (25,10))

sns.barplot(x = week_num, y = pd.Series(week_confirmed).diff().fillna(0), ax = ax1)
sns.barplot(x = week_num, y = pd.Series(week_deaths).diff().fillna(0), ax = ax2)
sns.barplot(x = week_num, y = pd.Series(week_recovered).diff().fillna(0), ax = ax3)


ax1.set_xlabel("WEEK NUMBER",color="red")
ax2.set_xlabel("WEEK NUMBER",color="red")
ax3.set_xlabel("WEEK NUMBER",color="red")

ax1.set_ylabel("NUMBER OF CONFIRMED CASES",color="green")
ax2.set_ylabel("NUMBER OF DEATH CASES",color="green")
ax3.set_ylabel("NUMBER OF RECOVERED CASES",color="green")

ax1.set_title("WEEKLY PROGRESS OF CONFIRMED CASES",color="blue")
ax2.set_title("WEEKLY PROGRESS OF DEATH CASES",color="blue")
ax3.set_title("WEEKLY PROGRESS OF RECOVERED CASES",color="blue")

plt.show()
```
