# COVID 19 in India

```python
covid_data
```

```python
covid_data.drop(["Province/State"],axis = 1,inplace=True)
```

```python
df = covid_data.copy()
df
```

```python
df.rename({"Country/Region":"Country"},axis=1,inplace=True)
```

```python
data_india = df[df["Country"] == "India"]
```

```python
data_india
```

```python
top_20_confirmed_cases = data_india.sort_values(["Confirmed"],ascending = False).head(20)
top_20_confirmed_cases
```

```python
top_20_recovered_cases = data_india.sort_values(["Recovered"],ascending = False).head(20)
top_20_recovered_cases
```

```python
top_20_deaths_cases = data_india.sort_values(["Deaths"],ascending = False).head(20)
top_20_deaths_cases
```

```python
plt.figure(figsize=(20,8))

plt.plot(data_india['ObservationDate'], data_india['Confirmed'] , label='Confirmed')
plt.plot(data_india['ObservationDate'], data_india['Recovered'] , label='Recovered')
plt.plot(data_india['ObservationDate'], data_india['Deaths'] , label='Deaths')

plt.xlabel('')
plt.ylabel('no of cases')
plt.title('Daily increase')

plt.legend()
plt.show()
```

```python
date_wise_india = data_india.groupby(["ObservationDate"]).agg({"Confirmed":"sum","Recovered":"sum","Deaths":"sum"})
```

```python
date_wise_india
```

```python
date_wise_india.iloc[-1]
```

## Total Active Cases in India

```python
Active_Cases_india = date_wise_india["Confirmed"].iloc[-1] - date_wise_india["Recovered"].iloc[-1] - date_wise_india["Deaths"].iloc[-1]
Active_Cases_india
```

## Total Closed Cases in India

```python
Closed_Cases_india = date_wise_india["Recovered"].iloc[-1] + date_wise_india["Deaths"].iloc[-1]
Closed_Cases_india
```

## Week wise in India

```python
date_wise_india["Week"] = date_wise_india.index.weekofyear
```

```python
date_wise_india
```

```python
week_num_india = []
week_confirmed_india = []
week_recovered_india = []
week_deaths_india = []
```

```python
w = 1
for i in list(date_wise_india["Week"].unique()):
    week_confirmed_india.append(date_wise_india[date_wise_india["Week"] == i]["Confirmed"].iloc[-1])
    week_recovered_india.append(date_wise_india[date_wise_india["Week"] == i]["Recovered"].iloc[-1])
    week_deaths_india.append(date_wise_india[date_wise_india["Week"] == i]["Deaths"].iloc[-1])
    week_num_india.append(w)
    w = w + 1
```

```python
week_num_india
```

```python
week_confirmed_india
```

```python
week_recovered_india
```

```python
week_deaths_india
```

```python
plt.plot(week_num_india,week_confirmed_india,color="violet",marker="o")
plt.plot(week_num_india,week_recovered_india,color="green",marker="x")
plt.plot(week_num_india,week_deaths_india,color="black",marker="*")

plt.xlabel("WEEKS",fontsize=15,color="red")
plt.ylabel("PROGRESS IN EACH CASES",fontsize=15,color="red")
plt.title("WEEKLY CASES",fontsize=20,color="blue")

plt.show()
```
