# 2. DATA CLEANING

```python
# retrieving data by using set and head:
# head is used for retrieving 24 datas

import numpy as np

set=pd.DataFrame(data.head(24),columns=["Index","Product_name","Product_price","Product_rating","Review"])  
set
```

# Checking for missing value by using isnull or notnull:

```python
# Checking for missing values by using isnull():

set["Review"].isnull()
```

```python
# Checking for missing values by using notnull():

set["Review"].notnull()
```

```python
# Total count of none values:

set.isnull().sum()
```

```python
# overall count for dataset:

set.count()
```

# Cleaning/filling missing data:

```python
# droping missing values:

set=pd.DataFrame(data.head(21),columns=["Index","Product_name","Product_price","Product_rating","Review"])
set
```

```python
# replacing values with some specific value

set.replace({"Good":1,"Bad":0})
```

# Data Frame methods:

```python
# slicing of row by using iloc[]:

set=set.iloc[0:11]
set
```

```python
# slicing of column by using iloc[]:

set=set.iloc[:,0:3]
set
```

```python
# assigning indexing pandas as dataframes:

set=pd.DataFrame(data,columns=["Product_name","Product_price","Product_rating","Review"],
                        index=["a","b","c","d"])
set
```

```python
# sorting by columns in ascending order:

set=pd.DataFrame(data.head(20),columns=["Product_name","Product_price","Product_rating","Review"])
set
set.sort_values(by=["Product_name"])
```

```python
# sorting columns by descending order:

set.sort_values(ascending=False, by=["Product_name"])
```
