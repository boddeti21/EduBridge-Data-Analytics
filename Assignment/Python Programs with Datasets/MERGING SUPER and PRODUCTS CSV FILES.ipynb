# Re-shaping data:

```python
# in the review column we can re-shape the data by categories into numbers Good=1,Bad=0

# Catergories Review data1:

import pandas as  pd
data1=pd.read_csv("C://Users//Tanuja//Desktop//DATA-SET//super.csv")
data1["Review"]=data1["Review"].map({"Good":1,"Bad":0}).astype(float)
data1
```

```python
#Finding the mean of the column having NaN

mean_value=data1["Review"].mean()
# Replace NaNs in column age with the
# mean of values in the same column

data1["Review"].fillna(value=mean_value, inplace=True)
print("Updated Dataframe:")
data1
```

```python
# in the review column we can re-shape the data by categories into numbers Good=1,Bad=0

# Catergories Review data2:

import pandas as  pd
data2=pd.read_csv("C://Users//Tanuja//Desktop//DATA-SET//products.csv")
data2["Review"]=data2["Review"].map({"Good":1,"Bad":0}).astype(float)
data2
```
